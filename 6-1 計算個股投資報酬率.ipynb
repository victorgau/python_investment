# 計算個股投資報酬率

## 載入需要的模組

```python
import pandas_datareader.data as web
```

## 設定資料的起始範圍

```python
start = "1990-01-01"
end = "2019-12-31"
```

## 讀取個股歷史資料

```python
df = web.DataReader('TSLA', 'yahoo', start, end)
```

# 計算夏普率 (Sharpe Ratio)

夏普率的定義：

$$夏普率=\frac{超額報酬平均值}{超額報酬標準差}$$

其中，

$$超額報酬 = 投資組合報酬 - 比較基準報酬$$

```python
import numpy as np
import pandas as pd
import pandas_datareader.data as web
```

```python
start = "2017-01-01"
end = "2017-12-31"
```

---

## SPY的Sharpe Ratio

```python
# 讀取從指定日期之後的股價資訊
df = web.DataReader("SPY", 'yahoo', start, end)
```

```python
dailyRet = df['Adj Close'].pct_change()
```

```python
#假設無風險利率為 4%
#假設一年有252個交易日
excessRet = dailyRet - 0.04/252

sharpeRatio = np.sqrt(252.0)*np.mean(excessRet)/np.std(excessRet)
```

```python
sharpeRatio
```

## 過去N年的Sharpe Ratio

```python
from datetime import datetime
from datetime import timedelta
```

```python
N = 3
```

```python
start = datetime.now() - timedelta(days=365*N)
```

```python
df = web.DataReader("SPY", 'yahoo', start)
```

```python
dailyRet = df['Adj Close'].pct_change()
```

```python
#假設無風險利率為 4%
#假設一年有252個交易日
excessRet = dailyRet - 0.04/252

sharpeRatio = np.sqrt(252.0)*np.mean(excessRet)/np.std(excessRet)
```

```python
sharpeRatio
```

### 載入需要的套件

```python
%matplotlib inline

import numpy as np
import pandas as pd
import pandas_datareader.data as web
import backtest
```

### 設定資料的範圍

```python
start = "2017-01-01"
end = "2017-12-31"
```

---

## SPY

```python
# 讀取從指定日期之後的股價資訊
df = web.DataReader("SPY", 'yahoo', start, end)
```

```python
df.head()
```

```python
dailyRet = df['Adj Close'].pct_change()
```

```python
cumRet = np.cumprod(1 + dailyRet)
```

### 計算 Maximum Drawdown

```python
backtest.calc_max_drawdown(cumRet)
```

### 進一步分析 Drawdown

```python
dd_series = backtest.to_drawdown_series(cumRet)
```

```python
dd_details = backtest.drawdown_details(dd_series)
```

```python
dd_details
```

### 同時計算 MaxDD 跟 MaxDDD

```python
backtest.DrawDownAnalysis(cumRet)
```

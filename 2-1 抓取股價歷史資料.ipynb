在這裡我們使用 pandas data-reader 來讀取股價歷史資料。

## 載入需要的模組

```python
from datetime import datetime
import pandas_datareader.data as web
```

## 設定資料的起始範圍

```python
start = datetime(2017, 1, 1)
end = datetime(2017, 12, 31)
```

## 讀取個股歷史資料

```python
df = web.DataReader('2330.tw', "yahoo", start, end)
```

# 下載特定的台股歷史股價，並存成檔案

```python
import pandas as pd
import pandas_datareader.data as web
```

## 抓取資料

```python
portfolio = ["2330.tw", "2317.tw", "2412.tw", "2454.tw"]
```

```python
start = "2015-01-01"
end = "2019-12-31"
```

```python
data = {}

for symbol in portfolio:
    try:
        data[symbol] = web.DataReader(symbol, "yahoo", start, end)
    except Exception as e:
        print("Error accesing symbol:", symbol, "==>", e.args)
```

```python
# 看看有幾筆資料
len(data.keys())
```

## 將股票序列化並存成檔案

```python
filename = 'twstockdata' 
```

```python
import pickle

with open(filename, 'wb') as f:
    pickle.dump(data, file=f)
```

## 將股票資料讀回來使用看看

```python
with open(filename, 'rb') as f:
    stocks = pickle.load(file=f)
```

```python
stocks.keys()
```

```python
%matplotlib inline
symbol = list(stocks.keys())[0]
stocks[symbol]['Close'].plot()
```
